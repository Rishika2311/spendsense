# SpendSense — EDA

Quick look at the unified CC + UPI dataset.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent / 'src'))
import pandas as pd
from spendsense.preprocess import load, add_time_features
df = add_time_features(load('../data/transactions.csv'))
df.head()
```

```python
df.groupby('source')['amount'].agg(['count','sum','mean'])
```

```python
df.groupby('category')['amount'].sum().sort_values(ascending=False).plot.bar(figsize=(9,4));
```

```python
df.assign(m=df['timestamp'].dt.to_period('M').astype(str)).groupby(['m','source'])['amount'].sum().unstack().plot(figsize=(9,4));
```
